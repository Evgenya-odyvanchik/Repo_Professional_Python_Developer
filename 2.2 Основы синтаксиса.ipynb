```python
# Задача
# Если пользователь ввёл N-ое число
# Вывести с учтом окончания:
#  3 питона
#  7 питона
#  15 питона
#  21 питона
#  23 питона
#  56 питона
# ипользуйте оператор остатка от деления %%!

python_count = int(input("Сколько питонов"))\

# если 11-19 - питонов
# если последняя цифра 1 - питон
# если последняя цифра 2-3-4 - питона
# иначе - питонов

last_digit = python_count % 10 

if 11 <= python_count <= 19:
    morphy = "питонов"
elif last_digit == 1:
    morphy = "питон"
elif 2 <= python_count <= 4:
    morphy = "питона"
else:
    morphy = "питонов"

print(python_count, morphy)
```

```python
# Задача: Из 150 выполнено 30 , сколько это в процентах?

tasks_all = 150
tasks_solved = 30
task_frac = tasks_solved / tasks_all 
print(task_frac)
# доля = выполненные задания / все задания
# доля = 30 / 150 = 0.2
# Это значит, что выполнено 0.2 (или 2/10) от всех заданий.


# Теперь умножим полученный результат на 100 и получим значение доли в процентах
# Чтобы перевести долю в проценты, умножаем на 100: 
# проценты = доля × 100
# проценты = 0.2 × 100 = 20%


tasks_all = 150
tasks_solved = 30
task_percent = tasks_solved / tasks_all * 100
print(task_percent)

# затем значение можно округлить :

tasks_all = 150
tasks_solved = 30
task_percent = tasks_solved / tasks_all * 100
task_percent_round = round(task_percent)
# округляем до целого
print(task_percent_round)
```

```python
# Процеты - расчёты
# Типичная задача при расчёте процентов - найти количество процентов от числа
# Задача: Если сумма скидки равна 20 % , а цена - 3000, то сколько мы сэкономим?

price = 3000
percent = 20
# Сначала приводим процент к дроби
fracture = percent / 100 #получится 0.2
# Затем умножаем цену на дробь
discounted = price * fracture
# Выводим результат
print(discounted)
```

```python
# Задача 2: Вычислить цену со скидкой , если известна цена и скидка

price = 3000
discount = 20
# Сначала приводим процент к дроби
fracture = percent / 100  #получится 0.2
# заетм вычитаем из цены сумму скидки
price_discounted = price - price * fracture
# выводим результат
print(price_discounted)

# Исходная цена: 3000 руб.
# Скидка: 20% (или 600 руб.)
# Цена со скидкой: 2400 руб.

# МОЖНО ЭТОТ КОД СОКРАТИТЬ
price = 3000
discount = 20

price_discounted = price * (1 - discount / 100)
print(price_discounted)

# Развернуты первый вариант:
fracture = discount / 100        # 0.2 (скидка в виде дроби)
price_discounted = price - price * fracture  # 3000 − 3000 × 0.2 = 3000 − 600 = 2400

# сокращённый вариант
price_discounted = price * (1 - discount / 100)  # 3000 × (1 − 0.2) = 3000 × 0.8 = 2400

```

```python
# Приведение вместо преобразования
# В примере выше мы преобразовали bool в int  но на самом деле нам даже не нужно делать это вручную
# При математических операциях Python сам превращает bool в нужный тип , это называется приведение (coercion), или имплицитным преобразованием

working_on_monday = True
working_on_tuesday = True
working_on_wednesday = False

working_days_total = working_on_monday + working_on_tuesday + working_on_wednesday #Писать по порядку нужно , не в столбик иначе выведет True

print(working_days_total)
```
