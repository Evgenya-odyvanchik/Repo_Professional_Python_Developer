```python
#У Дениса день рождения. Он купил килограмм печенья и позвал друзей, но они пришли без подарков. Теперь Денис хочет написать письмо и потребовать с каждого друга стоимость печенья. Посчитайте, сколько денег Денису должен вернуть каждый друг.
##Дорогой друг! Надеюсь, печенье понравилось, верни, пожалуйста, ____ Обратите внимание: данные списаны с чека, поэтому хранятся в строковом формате и предварительно их нужно будет подготовить. Когда закончите, округлите результат до целого числа (без точки) c помощью функции round().

cookies_kg_price_str = "860"
cookies_kg_price_int = int(cookies_kg_price_str)  # цена за кг печенья
friends_counts = 5 # количество друзей
personal_fee = cookies_kg_price_int / friends_counts

print(f"Дорогой друг! Надеюсь, печенье понравилось, верни, пожалуйста, {round(personal_fee)}!")
```

```python
#МНОГОСТРОЧНЫЕ ПРИНТЫ
#Чтобы код смотрелся лучше, мы можем переносить содержимое print по разным строчкам:
print("Однажды по дорожке "
      "Я шел к себе домой "
      "Смотрю и вижу: кошки "
      "Сидят ко мне спиной")

#МНОГОСТРОЧНЫЕ СТРОКИ
#Чтобы разместить строку на нескольких строчках , используются тройные одинарные (''') или тройные двойные (""") кавычки:
print('''Однажды по дорожке
Я шел к себе домой 
Смотрю и вижу: кошки
Сидят ко мне спиной''')
```

```python
#СЕПАРАТОРЫ
#Ели мы записываем в print несколько значений через запятую, то они выводятся через пробел:
print("раз", "два", "трис")
#Однако мы можем переопределить раздеоить СЕПАРАТОР(separator). Он указывается в круглых скобках функций print  через запятую после перечисления всех выводимых значений.
#print(данные, sep="здесь_указывается_разделитель")

#ПРИМЕР:
#Разделитель - союз "и" с пробелами:
print("раз", "два", "три", sep="и")

#Разделитель слэш(/)
print("раз", "два", "три", sep="/")

#Без разделителя
print("раз", "два", "три", sep="")

#Разделитель - пробел
print("раз", "два", "три", sep="")
```

```python
#ПЕРЕНОС СТРОК
#В Python, чтобы перейти на следующую строку, мы используем символ \n (от new line):
print("Однажды \n по дороге \n я шёл \n к себе домой")
#или
print("Однажды", "по", "дороге",  "я", "шёл",  "к", "себе", "домой", sep="\n")
#Пример еще
print("Alfa", "Bravo", "Charlie", "Delta", "Echo", sep="\n")
```

```python
#КОНЕЦ СТРОКИ
#Однако мы можем переопределить конец строки с помощью end. Чаще всего это делается для того, чтобы вывести несколько сообщений в одну строку:
print("Однажды ", end=" ")
print("По дорожке ", end=" ")
print("Я шел", end=" ")
print("К себе домой", end="")
```

```python
#Разделитель может быть длиной в несколько символов:
print("Однажды ", end="\n——-\n")
print("По дорожке ", end="\n——-\n")
print("Я шел", end="\n——-\n")
print("К себе домой")
```

```python
#КАВЫЧКИ И ЭКРАНИРОВАНИЕ
#Когда ввести ковычки внутри ковычек
#Введите в терминале "да" или "нет"
#Внутри одинарных кавычек используются двойные:
print('Введите в терминале "да" или "нет"')

#Внутри двойных кавычек — одинарные:
print("Ведите в терминале 'да' или 'нет'")

#Тройные кавычки:
print("""Введите в терминале "да" или "нет" """)
```
